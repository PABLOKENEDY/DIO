```python
# Cirando Variaval do Tipo String

frase = "Python Para Análise de Daods"
```

```python
# Todas Consulta de texte inicia-se em "0"
# Exibindo a Primeira Letra da String "Texto"
frase[0]
```

```python
# Exibindo Letras de 0 a 10
frase[0:10]
```

```python
# Exibindo Letras A Partir da 10 Letra
frase[10:]
```

```python
# Exibindo Letras de 0 a 18
frase[:18]
```

```python
# Análisando Quantas Vezes um palavra aparece no texto.
frase.count("Python")
```

```python
# mostrando todo o texto em letras minuscula
frase.lower()
```

```python
# mostrando todo o texto em letras Maiuscúla
frase.upper()
```

```python
# Anterando Uma Palavra Por Outra em Python
frase.replace("Python" , "Programando")
```
