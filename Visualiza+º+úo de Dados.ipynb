```python
import pandas as pd
```

```python
df = pd.read_excel("/content/drive/MyDrive/ProjetoDados.xlsx",)
```

```python
df["LojaID"].value_counts(ascending=False)
```

```python
# Gerando grafico de barras
df["LojaID"].value_counts(ascending=False).plot.bar()
```

```python
# Gerando grafico de barras horizontais
df["LojaID"].value_counts(ascending=False).plot.barh()
```

```python
# Gerando grafico Pizza
df["LojaID"].value_counts(ascending=False).plot.pie()
```

```python
df.groupby(df["Data"].dt.year)["Vendas"].sum().plot.pie()
```

```python
# Total de Vendas por Cidade
df["Cidade"].value_counts()
```

```python
# Alterando Titulo e Adicionando o nome dos 2 eixos
import matplotlib.pyplot as plt
df["Cidade"].value_counts().plot.barh(title="Total de Vendas Por Cidade")
plt.xlabel("Cidade")
plt.ylabel("Qtde")
```

```python
# Alterando Cor da barra
import matplotlib.pyplot as plt
df["Cidade"].value_counts().plot.barh(title="Total de Vendas Por Cidade", color="red")
plt.xlabel("Cidade")
plt.ylabel("Qtde")
```

```python
# Alterando Estilo
plt.style.use("ggplot")
```

```python
df.groupby(df["Data"].dt.year)["Vendas"].sum().plot()
plt.xlabel("Data")
plt.ylabel("Venda")
plt.legend()
```
