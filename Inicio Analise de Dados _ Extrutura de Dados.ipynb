### *Dicionáirios*

```python
# para Utilizar o Dicionario em Python Utilizamod {}
dc = {"Limão":15, "Pêra":10, "Biscoito":11, "Uva":0} # Dicionario trabalha com chaves e valores
```

```python
dc
```

```python
# Acessando um valor no dicionairio atraves de chave
dc["Uva"]
```

```python
# Atualizando valor de Um Elemento
dc["Uva"] = 12
dc
```

```python
# Retornando todas as chave adicionadas
dc.keys()
```

```python
# Retornando Valores do dicionário
dc.values()
```

```python
# Verificar no dicionário de exixte um valor caso não inserir
dc.setdefault("Maracuja", 10)
```
